Generate 1000 random pts(using numpy)

```python
import numpy as np
x=[]
x= np.random.randint(100,size=1000)
print(x)
```

Generating slope and intercept (random)

```python
y=[]
m=np.random.randint(5)
c=np.random.randint(100)
for i in x:
    e=np.random.randint(-100,100)
    y.append((m*i)+c+e)
print(y)
```

```python
len(y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
x_train=x[:750]
x_test=x[750:]
y_train=y[:750]
y_test=y[750:]
```

```python
plt.scatter(x_train,y_train)
```

gradient descent

```python
m1 = 0
c1 = 0
alpha = 0.0000001 
itr = 3000
n = float(len(x_train))
for i in range(itr): 
    y_pred = m*x_train + c1
    m2 = (-2/n)*sum(x_train*(y_train-y_pred))
    c2 = (-2/n)*sum(y_train- y_pred)  
    m1 = m1-alpha*m2  
    c1 = c1-alpha*c2  
    
```

```python
m1,c1 #5k
```

```python
y_pred = c1 + (m1*x_test)
```

```python
print("mae :",np.abs(y_test-y_pred).mean())
print("rmse :",(((y_test-y_pred)**2).mean())**0.5)
```

```python
plt.plot(x_test,y_pred,c="b")
plt.scatter(x_test,y_test,c="r")
```
